```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
col_names = [
    "OPERAT_AIRLN_IATA_CD",
	"SCHD_LEG_DEP_AIRPRT_IATA_CD",
	"OPERAT_FLIGHT_NBR",
	"SCHD_LEG_DEP_LCL_DT",
	"SCHD_LEG_ARVL_AIRPRT_IATA_CD",
	"SCHD_LEG_DEP_LCL_TMS",
	"SCHD_LEG_DEP_GMT_TMS",
	"SCHD_LEG_DEP_GMT_DT",
	"SCHD_LEG_ARVL_LCL_TMS",
	"SCHD_LEG_ARVL_LCL_DT",
	"SCHD_LEG_ARVL_GMT_TMS",
	"SCHD_LEG_ARVL_GMT_DT",
	"SCHD_LEG_BLOCK_MIN_CT",
	"MILE_GREAT_CIRCLE_DISTANC_QTY",
	"FLIGHT_LEG_TTL_AVAIL_SEAT_CT",
	"SCHD_AIRCAFT_EQUIP_CD",
	"CREW_EQUIP_CD",
	"FLEET_CD",
	"SUB_FLEET_CD",
	"BODYTYPE",
	"OPERATIONTYPE",
	"AVL_STA_REGION_ICAO_CD",
	"AVL_STA_MIRS_PRIORITY_CD",
	"AVL_STA_MIRS_PRIORITY_REGION_DESC",
	"INTERNATIONAL",
	"BANK_ID",
	"MKT_DAY_FREQUENCY",
	"MKT_WEEK_FREQUENCY",
	"MKT_DAY_DEP_SEQ",
	"BAG_SCAN_UTC_TMS",
	"BAG_SCAN_LCL_TMS",
	"SF_LOAD_TMS"
]
```

```python
df = pd.read_csv('AAdata.csv', delimiter='|')
df.columns = col_names
```

```python
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
datetime_columns = ["SCHD_LEG_DEP_LCL_TMS", "SCHD_LEG_DEP_GMT_TMS", 
                    "SCHD_LEG_ARVL_LCL_TMS", "SCHD_LEG_ARVL_GMT_TMS",
                   "BAG_SCAN_UTC_TMS", "BAG_SCAN_LCL_TMS"]

for col in datetime_columns:
    df[col] = pd.to_datetime(df[col], errors = 'coerce').dt.tz_localize('UTC', ambiguous='infer')
```

```python
df["SCHD_LEG_ARVL_GMT_TMS"].fillna(df["SCHD_LEG_ARVL_LCL_TMS"].dt.tz_convert("UTC"), inplace=True)
```

```python
df.isnull().sum()
```

```python
df[(df['OPERAT_FLIGHT_NBR'] == 940) & (df['SCHD_LEG_ARVL_AIRPRT_IATA_CD'] == 'BOS')]["SCHD_LEG_ARVL_GMT_TMS"].unique()
```

```python
dropped_cols = ['OPERAT_AIRLN_IATA_CD', 'SCHD_LEG_DEP_AIRPRT_IATA_CD', 'SCHD_LEG_DEP_LCL_DT', 'SCHD_LEG_DEP_LCL_TMS',
                'SCHD_LEG_DEP_GMT_DT', 'SCHD_LEG_ARVL_LCL_TMS','SCHD_LEG_ARVL_LCL_DT', 'SCHD_LEG_ARVL_GMT_DT',
                'SCHD_AIRCAFT_EQUIP_CD', 'CREW_EQUIP_CD', 'FLEET_CD', 'SUB_FLEET_CD', 'OPERATIONTYPE',
                'AVL_STA_REGION_ICAO_CD', 'AVL_STA_MIRS_PRIORITY_CD','AVL_STA_MIRS_PRIORITY_REGION_DESC', 
                'BAG_SCAN_LCL_TMS', 'SF_LOAD_TMS']
```

```python
working_df = df.drop(dropped_cols, axis = 1)
```

```python
working_df.shape
```

```python
working_df.head()
```

```python
working_df = working_df[working_df['BAG_SCAN_UTC_TMS'] < working_df['SCHD_LEG_DEP_GMT_TMS']]
```

```python
working_df.shape
```

```python
working_df["BAG_SCAN_BEFORE_DEPARTURE"] = (
    round((working_df['SCHD_LEG_DEP_GMT_TMS'] - working_df['BAG_SCAN_UTC_TMS']).dt.total_seconds() / 60, 2)
)
```

```python
working_df["BAG_SCAN_BEFORE_DEPARTURE"]
```

```python
working_df[working_df["BAG_SCAN_BEFORE_DEPARTURE"] < 0.1]
```

```python
working_df['SCHD_LEG_DEP_GMT_TMS'].min()
```

```python
working_df[working_df['OPERAT_FLIGHT_NBR'] == 2187]
```

```python
plt.figure(figsize=(10, 5))
plt.hist(working_df['BAG_SCAN_BEFORE_DEPARTURE'], bins=500, color='skyblue', edgecolor='black')
plt.xlim(0, 1000)
plt.xlabel('Bag Scan Before Departure (minutes)')
plt.ylabel('Frequency')
plt.title('Distribution of Bag Scan Times Before Departure')
```
